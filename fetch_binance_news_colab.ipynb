```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time

# Set up Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--window-size=1920x1080")
chrome_options.binary_location = r"C:\Program Files\Google\Chrome\Application\chrome.exe"
chrome_options.add_argument("--no-sandbox")  # Disable sandbox
chrome_options.add_argument("--disable-dev-shm-usage")  # Add this line

# Set up Chrome service using webdriver_manager
service = Service(ChromeDriverManager().install())

# Initialize the Chrome driver
driver = webdriver.Chrome(service=service, options=chrome_options)

# URL to scrape
url = "https://www.binance.com/en/support/announcement/list/49"

try:
    # Load the page
    driver.get(url)

    # Wait for JavaScript to render content (adjust time as needed)
    time.sleep(5)

    # Get the page source after JavaScript execution
    html = driver.page_source

    # Parse the HTML with BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')

    # Example: Find all announcement titles (adjust selector as needed)
    announcement_titles = soup.find_all('div', class_='css-1fqbqjb')  # Example class, replace with actual class

    # Print the announcement titles
    for title in announcement_titles:
        print(title.text)

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Close the driver
    driver.quit()


!git add fetch_binance_news_calab.ipynb
!git commit -m "Add Colab notebook"
!git push origin main
```
