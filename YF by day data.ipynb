```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
!pip install XlsxWriter

# Download data
btc = yf.download('BTC-USD', start='2020-04-11', end='2025-04-11')
ndaq = yf.download('NDAQ', start='2020-04-11', end='2025-04-11')

# Normalize prices to compare percentage change
btc_normalized = btc['Close'] / btc['Close'].iloc[0] * 100
ndaq_normalized = ndaq['Close'] / ndaq['Close'].iloc[0] * 100

# Plot
plt.figure(figsize=(10, 6))
plt.plot(btc_normalized, label='BTC/USD', color='orange')
plt.plot(ndaq_normalized, label='NDAQ', color='blue')
plt.title('BTC vs NDAQ Normalized Performance (Apr 2020 - Apr 2025)')
plt.xlabel('Date')
plt.ylabel('Normalized Price (% of Initial Value)')
plt.legend()
plt.grid()
plt.show()

# Save data to Excel
with pd.ExcelWriter('market_data.xlsx', engine='xlsxwriter') as writer:
    btc.to_excel(writer, sheet_name='BTC-USD')
    ndaq.to_excel(writer, sheet_name='NDAQ')
```
