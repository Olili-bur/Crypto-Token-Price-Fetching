```python
import yfinance as yf
import pandas as pd

# Editable UTC dates and tickers
start_date = "2025-03-19 16:34:00"
end_date = "2025-03-19 16:36:00"
tickers = ["BNB-USD","BTC-USD"]

# Convert strings to datetime objects for yfinance
start_dt = pd.to_datetime(start_date)
end_dt = pd.to_datetime(end_date)

# Fetch 1-minute data
data = yf.download(tickers, start=start_dt, end=end_dt, interval="1m", auto_adjust=True)

# Select Close prices and rename columns
df = data["Close"].rename(columns=lambda x: x.replace("-USD", "_Close"))

# Reindex to preserve original ticker order
ordered_columns = [ticker.replace("-USD", "_Close") for ticker in tickers]
df = df[ordered_columns]

# Ensure numeric and drop NaN
df = df.astype(float).dropna(how="all")

# Timezone fix for Excel
df.index = df.index.tz_localize(None)

# Add a row with token names at the top
token_names = [ticker.replace("-USD", "") for ticker in tickers]
df.loc["Token"] = token_names  # Add new row
df = df.reindex(["Token"] + list(df.index[:-1]))  # Move "Token" row to top

# Save to Excel
df.to_excel("doge_usd_1min.xlsx")
print("Data saved to 'doge_usd_1min.xlsx'")
```
